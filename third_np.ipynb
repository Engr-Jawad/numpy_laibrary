```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("====>   dumy data analysis  <====")
print("sales data shape : ",sales_data.shape)
print("\ndata of the ist three resturants :\n",sales_data[:3])
print("\nall resturant data with out id :\n",sales_data[:,1:])
```

#### total sales per year

```python
total_sales=np.sum(sales_data[:,1:],axis=0)  # column wise
print("total sales per year : ",total_sales)

total_sales=np.sum(sales_data[:,1:],axis=1) # row wise
print("toatal sales per year : ",total_sales)
# sum of total sales  
total_sales_all_rest=np.sum(total_sales)
print("total sales of all resturant combined : ",total_sales_all_rest)
```

```python
## minimum sales per resturant
minimum_sales = np.min(sales_data[:,1:],axis = 1)
print("minimum sales per resturant : ",minimum_sales)

## maximum sales per year
max_sales= np.max(sales_data[:,1:],axis = 0)
print("maximum sales per year : ",max_sales)

# average sales per resturant 
average_sale=np.mean(sales_data[:,1:],axis=1)
print("average sales per resturant : ",average_sale)
```

### cummulative sales of the resturants


```python
cum_sales = np .cumsum(sales_data[:,1:],axis = 1)
print("cummlative sales of the resturnat : \n",cum_sales)


# i print below the last columns of all resturant cummulataive sale 
cum_sales_total=cum_sales[:,3:]
print("\ntotal cum sale of the restuant : \n",cum_sales_total)
```

##  flotting graph using matplotlib

```python
plt.figsize=((10,4))
plt.plot(sales_data[:,1:])
plt.plot(np.mean(cum_sales,axis=0))
plt.title("sales per resturants ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid("True")
plt.show()
```

### vector addition and vector multiplication

```python
vec_one=np.array([1,2,3])
vec_two=np.array([3,2,1])
print("vector addition : ",vec_one + vec_two)
print("vector multiplication : ",vec_one * vec_two)
print("dot product : ",np.dot(vec_one,vec_two))
```

### find angle 

```python
angle = np.arccos(np.dot(vec_one,vec_two) / (np.linalg.norm(vec_one) * np.linalg.norm(vec_two)))
print("angle : ",angle)
```
