### creating array from list
<p> list and array are two diffrent concept in python 
<P> list are slower and np array are faster than list


```python
one_d=np.array([1,2,3,4,5])
print("one dimensional array : ",one_d)
two_d=np.array([[1,2,3],[3,4,5]])
print("two dimensional array : ",two_d)
```

### list vs np array

```python
list=[1,2,3]
print("list multiplication : ",list * 2)
# in numpy array element wise multiplication occured
np_array=np.array([1,2,3])
print("numpy array multiplictaion : ",np_array * 2)



start=time.time()
list=[i*3 for i in range (9000000)]
print("list operation time : ",time.time()-start)

startt=time.time()
np_arrayy=np.arange(9000000) * 3
print("numpy array operation time : ",time.time()-startt)
```

### creating array from scratch

```python
ones_array=np.ones((3,3))
print("array of ones : \n",ones_array)
print("\n")

zeros_array=np.zeros((3,3))
print("array of zeros : \n",zeros_array)

print("\n")
full_array=np.full((3,3),5)
print("full array  : \n",full_array)


random=np.random.random((3,3))
print("\nrandom array :  \n",random) 

sequance_array=np.arange(0,11,2)
print("\nsequance array : ",sequance_array)
```

### vector ,matrix  ,tensors
<p> tensors mean n dimensional array

```python
vector=np.array([1,2,3,4])
print("vector :",vector)

matrix=np.array([[1,2,3],
                 [4,5,6],
                 [7,4,2]])
print("\nmatrix : \n",matrix)

tensor=np.array([
    [[1,2],[3,4]],
    [[3,2],[6,2]],
    [[1,4],[9,0]]
])
print("tensors : \n",tensor)
```

### properties of array

```python
array=np.array([[1,2,3],
                [4,3,2],
                [3,2,1]])
print("shape : ",array.shape)
print("dimension : ",array.ndim)
print("size : ",array.size) # TOTAL NUMBER OF ELEMENT IN THE ARRAY 
print("data type : ",array.dtype)
```

### array reshaping

```python
array=np.arange(15)
print("orignal array  ",array)

reshaped_array=array.reshape(3,5)
print("\nreshaped array \n",reshaped_array)

flat=reshaped_array.flatten()
print("\nflatten the reshaped array : ",flat)


# raveled return vies instead of copy of the orignal array
raveled=reshaped_array.ravel()
print("\nraveled array : \n",raveled)


transpose = reshaped_array.T
print("\n transpose \n",transpose)






```
