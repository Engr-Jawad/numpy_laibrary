### find even number

### 

```python
arr = np.array([10, 15, 20, 25, 30, 35])
even_number = arr [ arr % 2 == 0]
print("even number : ",even_number)
```

### row and column sum

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
row_sum=np.sum(arr,axis=1)
print("row sum : ",row_sum)
column_sum=np.sum(arr,axis=0)
print("column sum : ",column_sum)
```

### matrix multiplication 

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("compaitability : ",A.shape == B.shape)
print("multiplication of matrix : ",np.dot(A,B))
```
