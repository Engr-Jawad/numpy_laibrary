# begginer level problems on numpy

```python
import numpy as np
import pandas as pd
```

```python
array=np.arange(11)
print("array from one to ten : ",array)

print("\n shape of array : ",array.shape)
print("\n size of array :",array.size)
```

```python
two_darray=np.random.random((3,3))
print("two dimensional array : \n",two_darray)
print("\nshap of the two d array : ",two_darray.shape)
print("\n size of the two d array : ",two_darray.size)
print("data type of the two d array  :",two_darray.dtype)
```

### creating array of zeros and ones

```python
zeros_array =np.zeros((5))
print("five  zeros array : ",zeros_array)
ones_array= np.ones((5))
print("five ones array : ",ones_array)
```

### creating array of even number from one to twenty

```python
even_array = np.arange(0,20,2)
print("even numbers array : ",even_array)
```

### Create an array from 1 to 12 and reshape it into a 3×4 matrix.

```python
array = np.arange(12)
print("original array  : ",array)

reshaped_array = array.reshape((3,4))
print("\nreshaped array : \n",reshaped_array)
```

### slicing concept

```python
array = np.arange(20)
print("orignal array " ,array)
print("\nslicing ist three element : ",array[0:3])
print("\nslicing last two element from array : ",array[18:20])
```

### addition and multiplication

```python
ist_array = np.array([2,3,6,9])
second_array = np.array([3,2,5,1])
print("addition of the two array : ",ist_array + second_array)

print("\n multiplication of two array : ",ist_array * second_array)
```

### Create an array and find:

maximum value
minimum value

```python
array = np.array([90,-4,30,3000,2000,-100,80000000])
print("maximum in the array : ",array.max())
print("minimum in the array : ",array.min())
```

### Mean and Sum
Calculate:

sum
mean
of an array

```python
array = np.array([5,6,4,3,77,88,33,99])
print("sum of the array : ",array.sum())
print("mean of the array : ",array.mean())
```

### 👉 Extract all elements greater than 5 from your array:

```python
array = np.array([5,6,4,3,77,88,33,99])
greater = array[array > 5]
print("greater element than 5  is : ",greater)
```
