## You are given exam scores of 5 students across 3 subjects (Math, Science, English).
Your task is to perform a series of analyses using NumPy.

```python
import numpy as np
```

```python
grades = np.array([
    [85, 90, 78],
    [92, 88, 84],
    [75, 70, 80],
    [88, 85, 89],
    [95, 93, 91]
])
# find mean of the score
mean_score_sub = np.mean(grades,axis=0)
print("mean score of the subject wise : ",mean_score_sub)
# maximum score in the subjects wise
max_score=np.max(grades,axis=0)
print("maximum score in the subject wise : ",max_score)
# minimum score in the subjects
min_score  = np.min(grades,axis=0)
print("minimum score subject wise: ",min_score)
```

## Compute for each student:

Total score (sum of all subjects)

Average score

```python
total_score = np.sum(grades,axis=1)
print("total score per student : ",total_score)
average_score = np.mean(grades,axis=1)
print(average_score)
```

## Find the student with the highest total score and their total.

```python
top_student_index = np.argmax(total_score)
top_student_score = np.max(total_score)
print(f"the index of the student is {top_student_index} and the scores equall to : {top_student_score}")
```

## The teacher decides to add 5 bonus points to all Math scores.

```python
grades[:,0] = grades[:,0] + 5
print(grades)
```

## Normalize scores (min-max scaling)



```python
min_values = np.min(grades,axis=0)
max_values =np.max(grades,axis=0)
normalized = (grades - min_values)/ ( max_values - min_values)
print("grades are normalixed : ",normalized.round(3))
```

Given an array, replace all values:

< 0 → 0
> 100 → 100
Do it without loops.

```python


arr1 = np.array([-10, 5, 120, 45, -3, 200, 67])
relplaced_with_zero = np.where(arr1 < 0 ,0,arr1)
print("elements are replaced which are less than zero : ",relplaced_with_zero)
relplaced_with_hundred = np.where(arr1 > 100 ,100,arr1)
print("relplaced_with_hundred : ",relplaced_with_hundred)
```

#### row wise normalization
