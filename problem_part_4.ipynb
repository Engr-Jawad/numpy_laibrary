### The Problem: Filtering and Normalizing a Matrix
Objective: Given a 2D NumPy array, filter out rows that contain any negative numbers, and then normalize the remaining rows so that their elements sum to 1. 
Requirements
Create a NumPy array of shape (4, 4) with random integers between -5 and 10 (inclusive).
Filter: Remove any entire row that contains at least one negative number.
Normalize: For the remaining rows, modify them in-place such that the sum of elements in each row is 1.0. 

```python
import numpy as np

```

```python
array =np.random.randint(-5,11,(4,4))
print("original array " ,array)
```

```python
condition = (array >= 0).all(axis=1)
# filter
non_negative= array[condition]
print("negative array remove from the matrix : \n",non_negative)
row_sum = np.sum(non_negative,axis=1,keepdims=True)
print("\nadding row wise ",row_sum)
normalized = non_negative / row_sum
print("\nnormalize matrix " ,normalized)
```
