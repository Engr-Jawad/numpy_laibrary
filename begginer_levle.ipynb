<br> Get the 3rd element
<br>Get last two elements
<br>Reverse the array

```python
arr = np.array([10, 20, 30, 40, 50])
print("third element of the array : ",arr[2])
print("last two element of the arra",arr[3:5])
reversed = arr[::-1]
print("array are reversed : ",reversed)
```

<br>Add both arrays

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("adding two array : ", a+b)
print("multiply both array",a * b)
print("dot product of the both array : ",np.dot(a,b))
```
