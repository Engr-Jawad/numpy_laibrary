### From a 4x4 array, extract all even rows and odd columns.

```python
matrix_five =np.random.randint(0,10,(4,4))
print("matrix : \n",matrix_five)
extraxt_even_rows = matrix_five[0:4:2]
print("extracting all even rows from the matrix : \n",extraxt_even_rows)
extract_odd_col=matrix_five[:,1:4:2]
print("extracting all odd columns : \n ",extract_odd_col)

result = matrix_five[0:4:2, 1:4:2]  # Rows 0,2 and Columns 1,3
print("Elements in even rows AND odd columns:")
print(result)
```

### Reshape a 1D array of 12 elements into a 3x4 matrix. Then flatten it back.

```python
arr = np.arange(0,12)
print("original array :  ",arr)
print("\nreshape it to 3 X 4 : \n",arr.reshape(3,4))
print("\nflatten it back :",arr.flatten())
```

### Stack two 2x2 arrays vertically (row-wise) and horizontally (column-wise

```python
arr = np .array([[1,2],
                 [3,4]])


arrr = np.array([[3,4],
                 [5,6]])
print("array one : \n",arr)
print("\narray two : \n",arrr)
vstack =np.vstack((arr,arrr))
print("row wise  : \n",vstack)
hstack = np.hstack((arr,arrr))
print("column wise : \n",hstack)
```

### 1. Normalize a 5x5 random matrix 

```python
matrix =np.random.randint(8,34,(6,6))
print("original matrix : \n",matrix)
mean_matrix = np.mean(matrix)
print("mean of the matrix : ",mean_matrix)
stddevi = np.std(matrix)
print("standard deviation of the matrix : ",stddevi)
normalized = (matrix - mean_matrix) / stddevi
print("normalized matrix : ",normalized.round(2))

```
